```python
labels = [i.strip() for i in "A, K, Q, J, T, 9, 8, 7, 6, 5, 4, 3, 2".split(',')]
```

```python
label_rank = {label : rank for rank, label in enumerate(labels)}
rank_label = {rank : label for rank, label in enumerate(labels)}
```

```python
def pattern_rank(label_pattern):
    temp = [label_rank[letter] for letter in label_pattern]
    return temp

def sort_label_pattern(label_pattern):
    temp = sorted([label_rank[letter] for letter in label_pattern])
    return temp

def return_pattern_label(sorted_pattern):
    return "".join([rank_label[rank] for rank in sorted_pattern])

def determine_kind(label_pattern):
    """
    Every hand is exactly one type. From strongest to weakest, they are:

        - 7 - Five of a kind, where all five cards have the same label: AAAAA
        - 6 - Four of a kind, where four cards have the same label and one card has a different label: AA8AA
        - 5 - Full house, where three cards have the same label, and the remaining two cards share a different label: 23332
        - 4 - Three of a kind, where three cards have the same label, and the remaining two cards are each different from any other card in the hand: TTT98
        - 3 - Two pair, where two cards share one label, two other cards share a second label, and the remaining card has a third label: 23432
        - 2 - One pair, where two cards share one label, and the other three cards have a different label from the pair and each other: A23A4
        - 1 - High card, where all cards' labels are distinct: 23456
    """
    kind = 0
    contains_labels = set(label_pattern)
    labels_count = {label : 0 for label in contains_labels}
    for i in contains_labels:
        labels_count[i] = label_pattern.count(i)
    # print(labels_count)
    if len(labels_count) == 1:
         kind = 7
    elif len(labels_count) == 5:
        kind = 1
    else:
        for i in contains_labels:
            if labels_count[i] == 4:
                kind = 6
            elif labels_count[i] == 3 and len(labels_count) == 2:
                kind = 5
            elif labels_count[i] == 3 and len(labels_count) == 3:
                kind = 4
            elif labels_count[i] == 2 and len(labels_count) == 3:
                kind = 3
            elif labels_count[i] == 2 and len(labels_count) == 4:
                kind = 2
            # else: 
            #     print("error finding kind")
            
            if kind > 0:
                break
    return kind
```

```python
labels2 =  [i.strip() for i in "A, K, Q, T, 9, 8, 7, 6, 5, 4, 3, 2, J".split(",")]
```

```python
def pattern_rank(label_pattern):
    temp = [label_rank[letter] for letter in label_pattern]
    return temp

def determine_kind2(label_pattern):
    """
    Every hand is exactly one type. From strongest to weakest, they are:

        - 7 - Five of a kind, where all five cards have the same label: AAAAA
        - 6 - Four of a kind, where four cards have the same label and one card has a different label: AA8AA
        - 5 - Full house, where three cards have the same label, and the remaining two cards share a different label: 23332
        - 4 - Three of a kind, where three cards have the same label, and the remaining two cards are each different from any other card in the hand: TTT98
        - 3 - Two pair, where two cards share one label, two other cards share a second label, and the remaining card has a third label: 23432
        - 2 - One pair, where two cards share one label, and the other three cards have a different label from the pair and each other: A23A4
        - 1 - High card, where all cards' labels are distinct: 23456

    Now, J cards can pretend to be whatever card is best for the purpose of 
    determining hand type; for example, QJJQ2 is now considered four of a kind. 
    However, for the purpose of breaking ties between two hands of the same
    type, J is always treated as J, not the card it's pretending to be: JKKK2 
    is weaker than QQQQ2 because J is weaker than Q.
    """
    kind = 0
    contains_J = False
    contains_labels = set(label_pattern)
    if 'J' in contains_labels and len(contains_labels) > 1:
        contains_J = True

    labels_count = {label : 0 for label in contains_labels}
    
    max_label = ""
    maxx = 0

    for i in contains_labels:
        labels_count[i] = label_pattern.count(i)
        if contains_J and i != 'J':
            if labels_count[i] >= maxx:
                maxx = labels_count[i]
                max_label = i

    # print(labels_count)
    if contains_J:
        label_pattern.replace('J', max_label)
        labels_count[max_label] += labels_count['J']
        labels_count.pop('J')
        contains_labels.remove('J')
    
    if len(labels_count) == 1:
         kind = 7
    elif len(labels_count) == 5:
        kind = 1
    else:
        for i in contains_labels:
            if labels_count[i] == 4:
                kind = 6
            elif labels_count[i] == 3 and len(labels_count) == 2:
                kind = 5
            elif labels_count[i] == 3 and len(labels_count) == 3:
                kind = 4
            elif labels_count[i] == 2 and len(labels_count) == 3:
                kind = 3
            elif labels_count[i] == 2 and len(labels_count) == 4:
                kind = 2
            # else: 
            #     print("error finding kind")
            
            if kind > 0:
                break
    return kind
```
