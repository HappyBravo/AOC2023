```python
input_data_file = "./input_day-1.txt"
```

```python
input_data = []
with open(input_data_file) as f:
    input_data = [line.strip() for line in f.readlines()]
```

# PART 1

```python
def go_direction(line, direction = 1):
    for i in line[::direction]:
        if i in map(str, range(10)):
            return i

def read_numbers(line):
    return go_direction(line) + go_direction(line, -1)
```

```python
sum([int(read_numbers(i)) for i in input_data])
```

```python
letters = "one, two, three, four, five, six, seven, eight, nine"
```

```python
letters = letters.split(",")
```

```python
letter_digit_map = {}
letter_digit_map = {letter.strip() : digit for letter, digit in zip(letters, range(1, 10))}
```

```python
letter_digit_map
```

```python
my_summ_arr = []
corr_summ_arr = []
```

```python
def replace_letters_to_digit(line, letters_digit_map = letter_digit_map):
    _line = []
    _line[:] = line[:]
    _line = "".join(_line)
    # if "zero" in _line:
    #     print("has 0")
    #     input()
    for letter in letters_digit_map.keys():
        _line = _line.replace(letter, str(letters_digit_map[letter]))
    return _line
```

```python
countt = 0
summ = 0
for line in input_data:
    new_line = replace_letters_to_digit(line)
    n = int(read_numbers(new_line))
    summ += n
    print(countt, line, new_line, n, summ)
    my_summ_arr.append(summ)
    countt+=1
    # input()
print(summ)
```
