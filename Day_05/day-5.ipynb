```python
def gib_destination_point(point, from_to_map): 
    # WILL RETURN THE CORRESPONDING DESTINATION POINT
    dest_point = point
    _diff = 0
    for [_src_start, _dest_start, _range] in from_to_map:
        if point >= _src_start and point < _src_start+_range:
            _diff = point - _src_start
            dest_point = _dest_start + _diff
    return dest_point
```
