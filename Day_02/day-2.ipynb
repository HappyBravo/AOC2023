```python
input_data_file = "./input_day-2.txt"
```

```python
input_data = []
with open(input_data_file) as f:
    input_data = [line.strip() for line in f.readlines()]
```

# PART 1

```python
req_red = 12
req_green = 13
req_blue = 14
```

```python
master_dict = {}
for _line in input_data[:]:
    print(_line)
    game_no, subsets = _line.split(':')
    print(game_no)
    sets = subsets.split(';')
    _inner = {}
    for set_no, cubes_count in enumerate(sets):
        cubes = [i.strip().split() for i in cubes_count.split(',')]
        colors_count = {}
        for countt, color in cubes:
            colors_count[color] = int(countt)
        _inner[set_no] = colors_count
        print(cubes)
        print(colors_count)
    print(_inner)
    master_dict[int(game_no.split()[1])] = _inner

```

```python
master_dict
```

```python
not_valid_game_nos = []
for game_no in master_dict:
    print(game_no)
    flag = 1
    for set_no in master_dict[game_no]:
        if flag == 0 :
            break
        print(master_dict[game_no][set_no])
        for color in master_dict[game_no][set_no]:
            if (color == 'red' and master_dict[game_no][set_no][color] > req_red) or \
               (color == 'green' and master_dict[game_no][set_no][color] > req_green) or \
               (color == 'blue' and master_dict[game_no][set_no][color] > req_blue):
                not_valid_game_nos.append(game_no)
                flag = 0
                print("---")
                break
```

```python
sum(master_dict.keys()) - sum(not_valid_game_nos)
```

# PART 2


```python
game_no_power_cubes = {}

for game_no in master_dict:
    print(game_no)
    # game_no_power_cubes[game_no] = {}
    game_no_power_cubes[game_no] = {'red' : 0, 'blue' : 0, 'green' : 0}

    for set_no in master_dict[game_no]:
        for color in master_dict[game_no][set_no]:
            if (master_dict[game_no][set_no][color] > game_no_power_cubes[game_no][color]):
                game_no_power_cubes[game_no][color] = master_dict[game_no][set_no][color]
       
    print(game_no_power_cubes[game_no])
    print("---")
```

```python
sum_power = 0
for i in game_no_power_cubes:
    powerr = 1
    for color in game_no_power_cubes[i]:
        powerr*=game_no_power_cubes[i][color]
    sum_power+= powerr
```
